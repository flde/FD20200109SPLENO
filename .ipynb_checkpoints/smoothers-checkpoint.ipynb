```r
library(mgcv)
library(mgcViz)
```

```r
set.seed(42)
```

# Use CellRank data

```r
x <- read.csv("pt.csv", header=FALSE)$V1
y <- read.csv("gene.csv", header=FALSE)$V1
w <- as.matrix(read.csv("lineages.csv", header=FALSE))
n = length(x)
w3 = w[,3] + abs(rnorm(n, 0, 0.01))
w3 <- w3 / mean(w3)
model <- gam(y ~ s(x, bs='cr', k=5), weights=w3)
o <- getViz(model)
o <- plot(sm(o, 1))
o <- o + l_fitLine(colour = "red") + l_rug(mapping = aes(x=x, y=y), alpha = 0.8) +
    l_ciLine(mul = 5, colour = "blue", linetype = 2) + 
    l_points(shape = 19, size = 1, alpha = 0.1) + scale_colour_brewer(palette="Blues")

models <- list()
models[[1]] <- model
models_plot <- list()
models_plot[[1]] <- o

for (n_sims in 1:10) {
    sims <- simulate(model, nsim=n_sims)
    
    xs <- rep(x, n_sims + 1)
    ys <- c(y, c(sims))
    ws <- c(w3, rep(1e-6, n * n_sims))
    ws <- (ws / mean(ws)) / (n_sims + 1)
    stopifnot(assertthat::are_equal(sum(w3), sum(ws)))
    
    model2 <-gam(ys ~ s(xs, bs='cr', k=5), weights=ws)
    o <- getViz(model2)
    o <- plot(sm(o, 1), title="x")
    o <- o + l_fitLine(colour = "red") + l_rug(mapping = aes(x=x, y=y), alpha = 0.8) +
        l_ciLine(mul = 5, colour = "blue", linetype = 2) + 
        l_points(shape = 19, size = 1, alpha = 0.1) + scale_colour_brewer(palette="Blues") +
        labs(title=paste0("#sims=", n_sims))
    models[[n_sims + 1]] <- model2
    models_plot[[n_sims + 1]] <- o
}
```

```r
models_plot
```

```r
for (i in 1:(n_sims + 1)) {
    if (i == 1)
        print("original")
    else
        print(paste0("n_sims=", i - 1))
    m <- models[[i]]
    print(mean(m$Vp))
    print(summary(m))
}
```

# Use gamSim

```r
set.seed(42)
n <- 200  # number of points
n_sims <- 1000  # number of simulations/stacks
dat <- gamSim(3, n=n)
```

```r
b <- gam(y ~ s(x1) + s(x2), data=dat)
summary(b)
```

```r
sims <- simulate(b, nsim=n_sims)
weights <- c(rep(0.99999, n), rep(1e-3, n * n_sims))
weights <- weights / mean(weights) / (n_sims + 1)
sum(weights)
```

```r
xs1 <- rep(dat$x1, n_sims + 1)
xs2 <- rep(dat$x2, n_sims + 1)
ys <- c(dat$y, sims)
# ys <- rep(dat$y, n_sims + 1)
```

```r
b2 <- gam(ys ~ s(xs1) + s(xs2), weights=weights)
summary(b2)
```

```r
mean(b$Vp)
```

```r
mean(b2$Vp)
```
