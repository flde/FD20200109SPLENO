# Import PAGA connectivities 

```r
paga <- read.csv("result/paga/connectivities.csv", row.names=1, check.names=FALSE)
paga_nacl <- read.csv("result/paga/connectivities_nacl.csv", row.names=1, check.names=FALSE)
paga_cpg <- read.csv("result/paga/connectivities_cpg.csv", row.names=1, check.names=FALSE)
paga_diff <- paga_cpg-paga_nacl
```

# PAGA connectivity heatmap

```r
paga <- paga[cell_type$cell_type_fine, cell_type$cell_type_fine]
paga <- paga[cell_type_fine_order, cell_type_fine_order]

paga_diff <- paga_diff[cell_type$cell_type_fine, cell_type$cell_type_fine]
paga_diff <- paga_diff[cell_type_fine_order, cell_type_fine_order]

paga_nacl <- paga_nacl[cell_type$cell_type_fine, cell_type$cell_type_fine]
paga_nacl <- paga_nacl[cell_type_fine_order, cell_type_fine_order]

paga_cpg <- paga_cpg[cell_type$cell_type_fine, cell_type$cell_type_fine]
paga_cpg <- paga_cpg[cell_type_fine_order, cell_type_fine_order]
```

```r
options(repr.plot.width=20, repr.plot.height=5)

hm_1 <- paga_conn_hm(paga_nacl, "PAGA connectivity (NaCl)", "Conect.", color_max=rev(brewer.pal(11,"RdBu"))[1])
hm_2 <- paga_conn_hm(paga_cpg, "PAGA connectivity (CpG)", "Conect.")
hm_3 <- paga_conn_hm(paga_diff, "PAGA connectivity diff.", "Conect. diff.", diff=TRUE)

hm_1 + hm_2 + hm_3 + patchwork::plot_layout(ncol=3, nrow=1)
```
