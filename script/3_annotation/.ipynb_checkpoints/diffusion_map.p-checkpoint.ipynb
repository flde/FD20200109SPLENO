# Scanpy diffusion map and pseudotime 

```python
import scanpy as sc
import pandas as pd

import numpy as np

import os
```

```python
# Scanpy settings  
sc.settings.verbosity=0
sc.settings.vector_friendly = False
sc.set_figure_params(figsize=(5, 5), )
```

```python
os.chdir('/research/peer/fdeckert/FD20200109SPLENO')
```

# Import data

```python
adata = sc.read_h5ad('data/object/qc.h5ad')
obs = pd.read_csv('data/object/components/meta.csv', index_col=0)
```

```python
adata = adata[adata.obs.index.isin(obs.index.tolist())]
adata.obs = obs
```

# Subset data

```python
adata = adata[adata.obs['cell_type'].isin(['MEP', 'Ery (1)', 'Ery (2)', 'Ery (3)', 'Ery (4)', 'Ery (5)', 'Ery (6)'])].copy()
```

# Pre-processing 

```python
sc.pp.filter_genes(adata, min_cells=3)
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)
sc.pp.scale(adata)
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=20, use_rep='X_pca')
```

# Diffusion map

```python
sc.tl.diffmap(adata)

adata.uns['iroot'] = np.flatnonzero(adata.obs['cell_type']=='MEP')[0]
sc.tl.dpt(adata, n_branchings=0, n_dcs=10)

sc.pl.diffmap(adata, color=['cell_type', 'msMeg1', 'pHb_RNA', 'dpt_pseudotime'], wspace=0.25)
```

# Save diffusion maps

```python
pd.DataFrame(adata.obsm['X_diffmap'], index=adata.obs.index).to_csv("data/object/components/diffusion_map.csv")
pd.DataFrame(adata.obs['dpt_pseudotime'], index=adata.obs.index).to_csv("data/object/components/diffusion_pseudotime.csv")
```
