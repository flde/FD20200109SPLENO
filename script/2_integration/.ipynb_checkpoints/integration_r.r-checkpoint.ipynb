# Parameter settings

```r
# Files 
so_file <- "data/object/qc.rds"

# Plotting Theme
ggplot2::theme_set(theme_global_set()) # From project global source()
```

# Import Seurat object

```r
so <- readRDS(so_file)
DefaultAssay(so) <- "RNA"
```

# Remove doublets 

```r
so <- subset(so, subset = doublet_class=="Singlet" & doublet_class_int=="Singlet")
```

# SCTransform 

```r
so <- SCTransform(so, variable.features.n=3000, vars.to.regress=c("pMt_RNA"), assay="RNA", verbose=FALSE)
so <- RunPCA(so, npcs=50, assay="SCT", verbose=FALSE)
```

```r
hvg <- VariableFeatures(so)
```

```r
write.csv(hvg, "data/object/hvg.csv", row.names=FALSE)
```

# Harmony integration 

```r
so <- RunHarmony(so, group.by.vars="sample_group", max.iter.harmony=20, plot_convergence=FALSE)
```

```r
so <- FindNeighbors(so, dims=1:50, k.param=30, reduction="harmony", verbose=FALSE)
so <- FindClusters(so, verbose=FALSE)
so <- RunUMAP(so, dims=1:50, n.neighbors=30, reduction="harmony", verbose=FALSE)
```

```r
options(repr.plot.width=22.5, repr.plot.height=20)

dplot_1 <- dplot(so, reduction="umap", group_by="seurat_clusters") + ggtitle("Louvaine (Harmony)")
dplot_2 <- dplot(so, reduction="umap", group_by="treatment") + scale_color_manual(values=color$treatment) + ggtitle("Treatment")
dplot_3 <- dplot(so, reduction="umap", group_by="sample_rep") + ggtitle("Replicate")
dplot_4 <- dplot(so, reduction="umap", group_by="tissue") + scale_color_manual(values=color$tissue) + ggtitle("Tissue")
dplot_5 <- dplot(so, reduction="umap", group_by="cc_phase_class") + scale_color_manual(values=color$cc_phase_class) + ggtitle("Cell cycle")
dplot_6 <- dplot(so, reduction="umap", group_by="label_fine_haemosphere") + scale_color_manual(values=color$label_fine_haemosphere) + ggtitle("Haemosphere label")
fplot_1 <- fplot(so, reduction="umap", features="doublet_score_log2") + scale_colour_viridis_c(option="inferno") + ggtitle("Log2(Doublet score)")
fplot_2 <- fplot(so, reduction="umap", features="pMt_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Percentage Mt")
fplot_3 <- fplot(so, reduction="umap", features="pHb_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Percentage Hb")
fplot_4 <- fplot(so, reduction="umap", features="pRb_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Percentage Rb")
fplot_5 <- fplot(so, reduction="umap", features="nCount_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("UMI count")
fplot_6 <- fplot(so, reduction="umap", features="nFeature_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Feature count")

dplot_comb <- dplot_1 + dplot_2 + dplot_3 + dplot_4 + dplot_5 + dplot_6 + fplot_1 + fplot_2 + fplot_3 + fplot_4 + fplot_5 + fplot_6 + plot_layout(ncol=3)
dplot_comb
```

# Seurat SCTransform integration 

```r
so <- SplitObject(so, split.by="sample_group")
```

```r
so <- so[c("NaCl_Rep2", "NaCl_Rep1", "CpG_Rep2", "CpG_Rep1")]
```

```r
options(future.globals.maxSize=3000*1024^2)
```

```r
features <- SelectIntegrationFeatures(object.list=so, nfeatures=3000, verbose=FALSE)
so <- PrepSCTIntegration(object.list=so, anchor.features=features, verbose=FALSE)
anchors <- FindIntegrationAnchors(object.list=so, anchor.features=features, normalization.method="SCT", verbose=FALSE)
so <- IntegrateData(anchorset=anchors, normalization.method="SCT", verbose=FALSE)
```

```r
so <- RunPCA(so, npcs=50, assay="integrated", verbose=FALSE)
so <- FindNeighbors(so, dims=1:30, k.param=15, verbose=FALSE)
so <- FindClusters(so, verbose=FALSE)
so <- RunUMAP(so, dims=1:30, n.neighbors=30, verbose=FALSE)
```

```r
options(repr.plot.width=22.5, repr.plot.height=20)

dplot_1 <- dplot(so, reduction="umap", group_by="treatment") + scale_color_manual(values=color$treatment) + ggtitle("Treatment")
dplot_2 <- dplot(so, reduction="umap", group_by="sample_rep") + ggtitle("Replicate")
dplot_3 <- dplot(so, reduction="umap", group_by="tissue") + scale_color_manual(values=color$tissue) + ggtitle("Tissue")
dplot_4 <- dplot(so, reduction="umap", group_by="label_fine_haemosphere") + scale_color_manual(values=color$label_fine_haemosphere) + ggtitle("Haemosphere label")
fplot_1 <- fplot(so, reduction="umap", features="doublet_score_log2") + scale_colour_viridis_c(option="inferno") + ggtitle("Log2(Doublet score)")
dplot_5 <- dplot(so, reduction="umap", group_by="doublet_class") + scale_color_manual(values=c("Doublet"="black", "Singlet"="light grey")) + ggtitle("Doublet class sample")
dplot_6 <- dplot(so, reduction="umap", group_by="doublet_class_int") + scale_color_manual(values=c("Doublet"="black", "Singlet"="light grey")) + ggtitle("Doublet class integrated")
fplot_2 <- fplot(so, reduction="umap", features="pMt_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Percentage Mt")
fplot_3 <- fplot(so, reduction="umap", features="pHb_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Percentage Hb")
fplot_4 <- fplot(so, reduction="umap", features="pRb_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Percentage Rb")
fplot_5 <- fplot(so, reduction="umap", features="nCount_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("UMI count")
fplot_6 <- fplot(so, reduction="umap", features="nFeature_RNA") + scale_colour_viridis_c(option="inferno") + ggtitle("Feature count")

dplot_comb <- dplot_1 + dplot_2 + dplot_3 + dplot_4 + fplot_1 + dplot_5 + dplot_6 + fplot_2 + fplot_3 + fplot_4 + fplot_5 + fplot_6 + plot_layout(ncol=3)
dplot_comb
```
