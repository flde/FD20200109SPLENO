# Parameter settings

```r
# Seurat files 
so_file <- "data/object/seurat_clean.rds"

# Plotting Theme
ggplot2::theme_set(theme_global_set()) # From project global source()
```

# Import Seurat 

```r
so <- readRDS(so_file)
```

# Split data bei treatment and tissue group 

```r
so_1 <- subset(so, subset = treatment == "NaCl" & tissue == "Progenitor")
so_2 <- subset(so, subset = treatment == "CpG" & tissue == "Progenitor")
```

# Cell cycle genes 

```r
cc_kowalczyk <- read.csv("cc_kowalczyk.csv")
cc_kowalczyk <- cc_kowalczyk[cc_kowalczyk$sig_population >= 5, ]
```

# Create SCseq Object

```r
sc_1 <- SCseq(expdata = GetAssayData(so_1, slot = "counts", assay = "RNA"))
sc_2 <- SCseq(expdata = GetAssayData(so_2, slot = "counts", assay = "RNA"))
```

# SCseq filter data

Filterdata can be run with LBatch and bmode to remove genes associated with batch effects. VarID workflow also allows regression of batch effect in pruneKnn with regNB. in that case the regNB parameter of the compNoise function needs to be TRUE as well. Additional variables can be regressed out using the regVar variable with a data.frame. The rownames correspond to cell id and columns to values e.g. cell cycle. 

```r
batch_1 <- split(rownames(so_1@meta.data), f = so_1$sample_name)
batch_2 <- split(rownames(so_2@meta.data), f = so_2$sample_name)
```

```r
sc_1 <- filterdata(sc_1, mintotal = 1500, minexpr = 5, minnumber = 5, CGenes = NULL, FGenes = cc_kowalczyk$gene, LBatch = batch_1, knn = 10)
sc_2 <- filterdata(sc_2, mintotal = 1500, minexpr = 5, minnumber = 5, CGenes = NULL, FGenes = cc_kowalczyk$gene, LBatch = batch_2, knn = 10)
```

# Filtered expression matrix

```r
expdata_1 <- getExpData(sc_1)
expdata_2 <- getExpData(sc_2)
```

# Create variables for regression 

```r
# Batch 
batch_1 <- so_1$sample_name
names(batch_1) <- colnames(so_1)
batch_1 <- batch_1[names(batch_1) %in% colnames(expdata_1)]

batch_2 <- so_2$sample_name
names(batch_2) <- colnames(so_2)
batch_2 <- batch_2[names(batch_2) %in% colnames(expdata_2)]
```

# Prune KNN and compute cluster with transition probabilities

```r
res_1   <- pruneKnn(expdata_1, large = TRUE, regNB = TRUE, knn = 10, alpha = NULL, no_cores = NULL, seed = 12345)
cl_1    <- graphCluster(res_1, pvalue = 0.01)
probs_1 <- transitionProbs(res_1, cl_1)

res_2   <- pruneKnn(expdata_2, large = TRUE, regNB = TRUE, knn = 10, alpha = NULL, no_cores = NULL, seed = 12345)
cl_2    <- graphCluster(res_2, pvalue = 0.01)
probs_2 <- transitionProbs(res_2, cl_2)
```

# Compute noise from corrected variance

```r
noise_1 <- compNoise(expdata_1, res_1, regNB = TRUE, pvalue = 0.01, no_cores = NULL)
noise_2 <- compNoise(expdata_2, res_2, regNB = TRUE, pvalue = 0.01, no_cores = NULL)
```

# Update SCseq object 

```r
sc_1 <- updateSC(sc_1, res = res_1, cl = cl_1, noise = noise_1, flo = .1)
sc_2 <- updateSC(sc_2, res = res_2, cl = cl_2, noise = noise_2, flo = .1)
```

# Save output

```r
saveRDS(sc_1, "data/object/components/raceid_progenitor_nacl_p.rds")
saveRDS(sc_2, "data/object/components/raceid_progenitor_cpg_p.rds")

saveRDS(res_1, "data/object/components/raceid_progenitor_nacl_p_res.rds")
saveRDS(res_2, "data/object/components/raceid_progenitor_cpg_p_res.rds")

saveRDS(cl_1, "data/object/components/raceid_progenitor_nacl_p_cl.rds")
saveRDS(cl_2, "data/object/components/raceid_progenitor_cpg_p_cl.rds")

saveRDS(probs_1, "data/object/components/raceid_progenitor_nacl_p_probs.rds")
saveRDS(probs_2, "data/object/components/raceid_progenitor_cpg_p_probs.rds")
```
