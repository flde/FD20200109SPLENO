```r
# Select positive end point dea
features <- lapply(seq_along(dea_marker_ep), function(i) {
    
    features <- dea_marker_ep[[i]][dea_marker_ep[[i]]$minimump_p_val<=0.01 & dea_marker_ep[[i]]$NaCl_avg_log2FC >= 1.25 & dea_marker_ep[[i]]$CpG_avg_log2FC >= 1.25, ][["gene"]]
    return(features[1:length(features)])
    
})

# Select unique end point features 
features <- lapply(features, function(i) {i[i%in%unlist(features)[table(unlist(features))==1]]})
```

```r
assosciation_mastp <- tradeseq_mastp[["assosciation"]]
assosciation_mastp[is.na(assosciation_mastp)] <- 1
assosciation_megp <- tradeseq_megp[["assosciation"]]
assosciation_megp[is.na(assosciation_megp)] <- 1
assosciation_eb <- tradeseq_eb[["assosciation"]]
assosciation_eb[is.na(assosciation_eb)] <- 1
```

```r
assosciation_mastp$mastp <- assosciation_mastp$pvalue_1 <= 0.01
assosciation_megp$megp <- assosciation_megp$pvalue_1 <= 0.01
assosciation_eb$eb <- assosciation_eb$pvalue_1 <= 0.01
```

```r
assosciation_mastp <- assosciation_mastp[, "mastp", drop=FALSE]
assosciation_megp <- assosciation_megp[, "megp", drop=FALSE]
assosciation_eb <- assosciation_eb[, "eb", drop=FALSE]
```

```r
assosciation <- list(assosciation_mastp, assosciation_megp, assosciation_eb)
assosciation <- Reduce(function(x, y) merge(x, y, by=0, all=TRUE), assosciation)
assosciation[is.na(assosciation)] <- FALSE
```

```r
so_eb_nacl <- AddMetaData(so_eb_nacl, dpt_pseudotime_eb_nacl)
so_eb_cpg <- AddMetaData(so_eb_cpg, dpt_pseudotime_eb_cpg)
```

```r
options(repr.plot.width=7.5, repr.plot.height=2.5)

fplot_1 <- fplot(so_eb_nacl, reduction="umap", features="dpt_pseudotime") + ggtitle("EB (NaCl)") + scale_color_viridis(option="G")
fplot_2 <- fplot(so_eb_cpg, reduction="umap", features="dpt_pseudotime") + ggtitle("EB (CpG)") + scale_color_viridis(option="G")
 
fplot_1 + fplot_2 + plot_layout(ncol=2)
```

```r
counts_eb_nacl <- GetAssayData(so_eb_nacl, assay="RNA", slot="counts")
genes_eb_nacl <- rownames(counts_eb_nacl[rowSums(counts_eb_nacl>5)>10, ])

counts_eb_cpg <- GetAssayData(so_eb_cpg, assay="RNA", slot="counts")
genes_eb_cpg <- rownames(counts_eb_cpg[rowSums(counts_eb_cpg>5)>10, ])
```

```r
genes_eb <- unique(genes_eb_nacl, genes_eb_cpg)
```

```r
tradeseq_eb_nacl <- tradeseq_workflow(so_eb_nacl, dpt_pseudotime_eb_nacl, suffix="_eb_nacl", genes=genes_eb, compute=FALSE)
tradeseq_eb_cpg <- tradeseq_workflow(so_eb_cpg, dpt_pseudotime_eb_cpg, suffix="_eb_cpg", genes=genes_eb, compute=FALSE)
```

```r
assosciation_eb_nacl <- tradeseq_eb_nacl[["assosciation"]]
assosciation_eb_nacl[is.na(assosciation_eb_nacl)] <- 1
assosciation_eb_nacl$eb_nacl <- assosciation_eb_nacl$pvalue_1 <= 0.01 & assosciation_eb_nacl$meanLogFC >= 0.25
assosciation_eb_nacl <- assosciation_eb_nacl[, "eb_nacl", drop=FALSE]

assosciation_eb_cpg <- tradeseq_eb_cpg[["assosciation"]]
assosciation_eb_cpg[is.na(assosciation_eb_cpg)] <- 1
assosciation_eb_cpg$eb_cpg <- assosciation_eb_cpg$pvalue_1 <= 0.01  & assosciation_eb_cpg$meanLogFC >= 0.25
assosciation_eb_cpg <- assosciation_eb_cpg[, "eb_cpg", drop=FALSE]
```

```r
assosciation <- list(assosciation_eb_nacl, assosciation_eb_cpg)
assosciation <- Reduce(function(x, y) merge(x, y, by=0, all=TRUE), assosciation)
assosciation[is.na(assosciation)] <- FALSE
assosciation <- column_to_rownames(assosciation, var="Row.names")
```

```r
options(repr.plot.width=5, repr.plot.height=5)

upset(data=assosciation[, c("eb_nacl", "eb_cpg")], intersect=list("eb_nacl", "eb_cpg"), 
      min_size=0,
      width_ratio=0.125) + theme_global_set()
```

```r
fitgam_eb_nacl <- tradeseq_eb_nacl[["fitgam"]]
fitgam_eb_cpg <- tradeseq_eb_cpg[["fitgam"]]
```

## Spearman correlation 

```r
genes <- rownames(assosciation[assosciation$eb_nacl==TRUE | assosciation$eb_cpg==TRUE, ])
```

```r
cnt_nacl <- GetAssayData(so_eb_nacl, assay="RNA", slot="data")
cnt_cpg <- GetAssayData(so_eb_cpg, assay="RNA", slot="data")
```

```r
cnt_nacl <- cnt_nacl[which(rownames(cnt_nacl) %in% genes), order(so_eb_nacl$dpt_pseudotime)]
cnt_cpg <- cnt_cpg[which(rownames(cnt_cpg) %in% genes), order(so_eb_cpg$dpt_pseudotime)]
```

```r
options(repr.plot.width=5, repr.plot.height=175*0.15)

n_bins <- 100

cnt_smooth <- cbind(
    
    predictSmooth(fitgam_eb_nacl, gene=rownames(assosciation[assosciation$eb_nacl==TRUE & assosciation$eb_cpg==FALSE, ]), nPoints=n_bins, tidy=FALSE), 
    predictSmooth(fitgam_eb_cpg, gene=rownames(assosciation[assosciation$eb_nacl==TRUE & assosciation$eb_cpg==FALSE, ]), nPoints=n_bins, tidy=FALSE)

)

cnt_smooth <- t(scale(t(cnt_smooth)))
breaks <- seq(-max(abs(cnt_smooth)), +max(abs(cnt_smooth)), by = 1)

hm_nacl <- pheatmap(
    
    cnt_smooth[, 1:n_bins],
    main="NaCl", 
    cluster_cols=FALSE,
    show_rownames=TRUE, 
    show_colnames=FALSE, 
    border_color=NA, 
    breaks=breaks 

)


hm_cpg <- pheatmap(
    
    cnt_smooth[, n_bins+1:n_bins/2],
    main="CpG",
    cluster_cols=FALSE,
    show_rownames=TRUE, 
    show_colnames=FALSE, 
    border_color=NA, 
    breaks=breaks

)

hm_nacl + hm_cpg 
```
