# Parameter settings

```r
compute <- FALSE

# Files 
so_file <- "data/object/seurat.rds"
cluster_file <- "data/object/components/cluster/scvi_snn_res.0.8"

# Plotting Theme
ggplot2::theme_set(theme_global_set()) # From project global source()
```

# Load Seurat 

```r
so <- readRDS(so_file)
```

# Load cluster  

```r
cluster <- read.table("data/object/components/cluster/scvibatch_snn_res.0.8.txt", row.names = 1, header = TRUE, sep = "\t")
```

## Transfer cluster to seurat meta data 

```r
so <- AddMetaData(so, cluster)
```

# Compute cluster marker 

```r
Idents(so) <- so$`scvi_snn_res.0.8`
if(compute) {
    cluster_deg <- FindAllMarkers(so, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25)
    saveRDS(cluster_deg, "result/cluster_deg.rds")
} else {
    cluster_deg <- readRDS("result/cluster_deg.rds")
}
```
