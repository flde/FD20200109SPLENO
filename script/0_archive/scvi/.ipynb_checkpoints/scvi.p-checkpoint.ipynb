# Loading and preparing data 

```python
import scvi
import scanpy as sc

import pandas as pd
import numpy as np
import scipy

import os
import sys

sc.set_figure_params(figsize=(5, 5))
```

# Import custom modules

```python
sys.path.append('/research/peer/fdeckert/scFacility/script/')
import dirFacility
```

# Import AnnData object

```python
adata = dirFacility.dir2adata('/research/peer/fdeckert/FD20200109SPLENO/data/object/seurat/', assay='RNA', slot='counts')
```

```python
adata.layers['counts'] = adata.X.copy() # preserve counts
```

# Load highly variable features

SCVI performs well if there are more cells than features. To be consistant it would make sense using SCTransform varialb genes for all workflows.

```python
hvf_1 = pd.read_csv('data/object/seurat_sct_nacl/assay/SCT/variable_features.csv').iloc[:, 0].tolist()
hvf_2 = pd.read_csv('data/object/seurat_sct_cpg/assay/SCT/variable_features.csv').iloc[:, 0].tolist()

hvf = list(set(hvf_1) | set(hvf_2))
```

```python
adata = adata[:, adata.var_names.isin(hvf)].copy()
```

# Register model

```python
adata=scvi.data.setup_anndata(
    adata,
    layer='counts',
    batch_key='sample_rep', 
    categorical_covariate_keys=['treatment'], 
    continuous_covariate_keys=['msCC_diff_RNA'], 
    copy=True
)
```

# Creating and training a model 

```python
model=scvi.model.SCVI(
    adata, 
    n_latent=10, 
    n_hidden=128, 
    n_layers=2, 
    dispersion="gene"
)
```

```python
model.train()
```

```python
model.save('data/object/seurat/model/', overwrite=True)
```

# Update anndata object with scvi results 

```python
# scvi normalized counts 
adata.layers['data']=model.get_normalized_expression(library_size=10e4)
```

```python
adata.obsm['latent']=model.get_latent_representation()
```

# Leiden cluster and UMAP

```python
sc.pp.neighbors(adata, use_rep="latent")
sc.tl.leiden(adata)
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color=["leiden", "tissue", "cc_phase_class", "treatment", "main_labels", "fine_labels"], frameon=False, ncols=3)
```

# Store results 

```python
adata.write('data/object/seurat/andata.h5ad')
```

```python
dirFacility.adata2dir(adata, 'data/object/seurat/', assay='SCVI', slot='data', build_dir=False, overwrite=True)
```
