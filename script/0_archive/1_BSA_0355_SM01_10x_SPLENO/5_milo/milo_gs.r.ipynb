# Convert Seurat to Milo object

```r
# Get components
counts <- GetAssayData(so, assay="RNA", slot="counts")
meta <- so@meta.data
latent <- so@reductions$latent@cell.embeddings
umap <- so@reductions$umap@cell.embeddings
```

```r
# Milo object
milo <- miloR::Milo(sce)
```

## infection Baseline vs CpG 

```r
milo_cache <- FALSE
```

```r
milo_tl <- function(milo, so, k) {
    
    # Design matrix 
    design_df <- so@meta.data[,c ("sample_group", "infection"), drop=FALSE] %>% dplyr::distinct()
    rownames(design_df) <- design_df$sample_group
    
    infection <- as.character(design_df$infection)
    design <- model.matrix(~0+infection, data=design_df)
    colnames(design) <- gsub("infection", "", colnames(design))

    # Run milo
    res <- testNhoods(milo, design, design_df, reduced.dim="latent", model.contrasts="CpG-Baseline", fdr.weighting="graph-overlap", norm.method="TMM") %>% as.data.frame()
    
    # P-value hist
    pvalue_hist_1 <- ggplot(res, aes(PValue)) + geom_histogram(bins=50) + ggtitle(paste("Baseline vs CpG\n k", milo@.k)) + theme_global_set(1)

    # Volcano plot 
    vp_1 <- ggplot(res, aes(logFC, -log10(SpatialFDR))) + geom_point() + geom_hline(yintercept=1) + ggtitle(paste("Baseline vs CpG\n k", milo@.k)) + theme_global_set(1)
    
    # Annotate hoods
    res <- annotateNhoods(milo, res, coldata_col="leiden_cell_type_main")
    
    # Factor names 
    res$leiden_cell_type_main <- factor(res$leiden_cell_type_main, levels=rev(levels(so$leiden_cell_type_main))) %>% droplevels()
    res <- res[order(res$leiden_cell_type_main), ]
    
    # Make names for beeswarm 
    res$leiden_cell_type_main <- make.names(res$leiden_cell_type_main)
    
    # Beeswarm 
    if(any(res$SpatialFDR<=1.0)) {
        
        dabp_1 <- suppressMessages(plotDAbeeswarm(res, group.by="leiden_cell_type_main", alpha=1) + scale_color_gradient2(low=color$infection[1], high=color$infection[2]) + xlab("") + ggtitle(paste("Baseline vs CpG\n k", milo@.k)) + theme_global_set(1))
        
    } else {
        
        dabp_1 <- suppressMessages(plotDAbeeswarm(res, group.by="leiden_cell_type_main", alpha=1) + xlab("") + ggtitle(paste("k", milo@.k)) + theme_global_set(1))
        
    }
    
    
    # Cobine result table
    res_da <- list(res)
    
    # Combine all
    resl <- list(pvalue_hist=pvalue_hist_1, vp=vp_1, dabp=dabp_1, res_da=res_da)
    
    return(resl)
    
    print("Done")
    
}
```

```r
if(!milo_cache) {
    
    resl <- lapply(seq_along(k_range), function(k) milo_tl(milo=milo[[k]], so=so, k=k_range[k]))
    saveRDS(resl, "data/BSA_0355_SM01_10x_SPLENO/ANALYSIS/milo/resl.rds")
                  
} else {
    
    resl <- readRDS("data/BSA_0355_SM01_10x_SPLENO/ANALYSIS/milo/resl.rds")
    
}
```
