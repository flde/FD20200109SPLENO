```r
so <- readRDS("data/object/raw_filter_label.rds")
so <- subset(so, subset=qc_class=="fail")
so <- subset(so, subset=pMt_RNA <= 15)
table(so$treatment)
```

```r
so <- NormalizeData(so, assay="RNA")
so <- ScaleData(so, assay="RNA", verbose=FALSE)
so <- RunPCA(so, npcs=50, assay="RNA", features=rownames(so), verbose=FALSE)
so <- FindNeighbors(so, dims=1:30, k.param=20, verbose=FALSE)
so <- FindClusters(so, verbose=FALSE, resolution=1, algorithm=1, group.singletons=FALSE)
so <- RunUMAP(so, dims=1:50, n.neighbors=30, verbose=FALSE)
```

```r
options(repr.plot.width=25, repr.plot.height=5)

p_1 <- dplot(so, reduction="umap", group_by="treatment") + scale_color_manual(values=color$treatment)
p_2 <- dplot(so, reduction="umap", group_by="label_fine_haemosphere") + scale_color_manual(values=color$label_fine_haemosphere)
p_3 <- dplot(so, reduction="umap", group_by="doublet_class")
p_4 <- fplot(so, reduction="umap", features="pMt_RNA")
p_5 <- fplot(so, reduction="umap", features="Bmp4")

p_1 + p_2 + p_3 + p_4 + p_5 + plot_layout(ncol=5)
```
