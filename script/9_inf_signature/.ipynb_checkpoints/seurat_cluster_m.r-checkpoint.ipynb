# Sub cluster 

```r
so_m <- subset(so, subset=cell_type_fine %in% c("cMo (1)", "cMo (2)", "ncMo (1)", "ncMo (2)", "cDC2 (1)", "cDC2 (2)", "cDC2 (3)", "RPMP", "RPM"))
```

```r
so_m <- SplitObject(so_m, split.by="sample_group")
so_m <- so_m[c("NaCl_Rep1", "NaCl_Rep2", "CpG_Rep1", "CpG_Rep2")]

so_m <- lapply(so_m, function(so) {
    
    so <- SCTransform(so, return.only.var.genes=TRUE, variable.features.n=4000, vars.to.regress=c("msCC_diff_RNA"), verbose=FALSE)
    so <- RunPCA(so, npcs=50, assay="SCT", verbose=FALSE)
    so <- FindNeighbors(so, dims=1:50, k.param=30, verbose=FALSE)
    so <- FindClusters(so, verbose=FALSE, resolution=1, algorithm=1, group.singletons=TRUE)
    so <- RunUMAP(so, dims=1:50, n.neighbors=30, verbose=FALSE)
    
    return(so)
    
}
                )
```

```r
options(repr.plot.width=20, repr.plot.height=5)

dplot_1 <- dplot(so_m[[1]], reduction="umap", group_by="cell_type_fine") + scale_color_manual(values=unlist(color$cell_type_fine)[names(unlist(color$cell_type_fine)) %in% so_m[[1]]$cell_type_fine])
dplot_2 <- dplot(so_m[[2]], reduction="umap", group_by="cell_type_fine") + scale_color_manual(values=unlist(color$cell_type_fine)[names(unlist(color$cell_type_fine)) %in% so_m[[2]]$cell_type_fine])
dplot_3 <- dplot(so_m[[3]], reduction="umap", group_by="cell_type_fine") + scale_color_manual(values=unlist(color$cell_type_fine)[names(unlist(color$cell_type_fine)) %in% so_m[[3]]$cell_type_fine])
dplot_4 <- dplot(so_m[[4]], reduction="umap", group_by="cell_type_fine") + scale_color_manual(values=unlist(color$cell_type_fine)[names(unlist(color$cell_type_fine)) %in% so_m[[4]]$cell_type_fine])

dplot_comb <- dplot_1 + dplot_2 + dplot_3 + dplot_4 + plot_layout(ncol=4)
dplot_comb
```

```r
integrate_workflow <- function(so) {
    
    features <- SelectIntegrationFeatures(object.list=so, nfeatures=8000)
    so <- PrepSCTIntegration(object.list=so, anchor.features=features)
    anchors <- FindIntegrationAnchors(object.list=so, normalization.method="SCT", anchor.features=features, verbose=FALSE)
    so <- IntegrateData(anchorset=anchors, normalization.method="SCT", verbose=FALSE)
    
    return(so)
    
}
```

```r
so_m_int_nacl <- integrate_workflow(so_m[c("NaCl_Rep1", "NaCl_Rep2")])
so_m_int_cpg <- integrate_workflow(so_m[c("CpG_Rep1", "CpG_Rep2")])
```

```r
dim_workflow <- function(so) {
    
    so <- RunPCA(so, npcs=30, assay="integrated", verbose=FALSE)
    so <- FindNeighbors(so, dims=1:10, k.param=30, verbose=FALSE)
    so <- FindClusters(so, verbose=FALSE, resolution=1, algorithm=1, group.singletons=TRUE)
    so <- RunUMAP(so, dims=1:10, n.neighbors=30, verbose=FALSE)
    
    return(so)
}
```

```r
so_m_int_nacl <- dim_workflow(so_m_int_nacl)
so_m_int_cpg <- dim_workflow(so_m_int_cpg)
```

```r
options(repr.plot.width=10, repr.plot.height=5)

dplot_1 <- dplot(so_m_int_nacl, reduction="umap", group_by="cell_type_fine") + scale_color_manual(values=unlist(color$cell_type_fine)[names(unlist(color$cell_type_fine)) %in% so_m_int_nacl$cell_type_fine])
dplot_2 <- dplot(so_m_int_cpg, reduction="umap", group_by="cell_type_fine") + scale_color_manual(values=unlist(color$cell_type_fine)[names(unlist(color$cell_type_fine)) %in% so_m_int_cpg$cell_type_fine])

dplot_comb <- dplot_1 + dplot_2 + plot_layout(ncol=2)
dplot_comb
```
