# Import

```r
# Seurat object
so <- readRDS(so_file)
```

```r
# Gene sets
gene_set_hallmark <- msigdbr(species="Mus musculus", category="H")
gene_set_hallmark <- split(x=gene_set_hallmark$gene_symbol, f=gene_set_hallmark$gs_name)

gene_set_reactome <- msigdbr(species="Mus musculus", category="C2", subcategory="CP:REACTOME") 
gene_set_reactome <- split(gene_set_reactome, x=gene_set_reactome$gene_symbol, f=gene_set_reactome$gs_name)

gene_set_go <- msigdbr(species="Mus musculus", category="C5", subcategory="GO:BP")
gene_set_go <- split(x=gene_set_go$gene_symbol, f=gene_set_go$gs_name)

gene_sets_inf_ab <- unique(
    c(
        gene_set_reactome$REACTOME_INTERFERON_ALPHA_BETA_SIGNALING, 
        gene_set_hallmark$HALLMARK_INTERFERON_ALPHA_RESPONSE, 
        gene_set_go$GOBP_CELLULAR_RESPONSE_TO_INTERFERON_ALPHA, 
        gene_set_go$GOBP_RESPONSE_TO_INTERFERON_ALPHA, 
        gene_set_go$GOBP_CELLULAR_RESPONSE_TO_INTERFERON_BETA, 
        gene_set_go$GOBP_RESPONSE_TO_INTERFERON_BETA, 
        gene_set_go$GOBP_RESPONSE_TO_TYPE_I_INTERFERON, 
        gene_set_go$GOBP_POSITIVE_REGULATION_OF_TYPE_I_INTERFERON_MEDIATED_SIGNALING_PATHWAY, 
        gene_set_go$GOBP_REGULATION_OF_TYPE_I_INTERFERON_MEDIATED_SIGNALING_PATHWAY
    )
)

gene_sets_inf_g <- unique(
    c(
        gene_set_reactome$REACTOME_INTERFERON_GAMMA_SIGNALING, 
        gene_set_hallmark$HALLMARK_INTERFERON_GAMMA_RESPONSE, 
        gene_set_go$GOBP_CELLULAR_RESPONSE_TO_INTERFERON_GAMMA, 
        gene_set_go$GOBP_RESPONSE_TO_INTERFERON_GAMMA
    )
)

# In vitro data set 
dea_inf_b_mac <- readRDS("result/inf_signature/dea_inf_beta_mac.rds")
genes_inf_b_mac <- unique(do.call("rbind", dea_inf_b_mac)$gene)

# In vitro data set 
dea_inf_g_mac <- readRDS("result/inf_signature/dea_inf_gamma_mac.rds")
genes_inf_g_mac <- unique(do.call("rbind", dea_inf_g_mac)$gene)

# Combine 
inf_ab_gens <- unique(c(gene_sets_inf_ab, genes_inf_b_mac))
inf_g_gens <- unique(c(gene_sets_inf_g, genes_inf_g_mac))
```

```r
inf_genes <- gene_sets_inf_ab
```

# Overall DEA in combined monocytes and RPM

```r
so_m <- subset(so, subset=cell_type_fine %in% c("cMo (1)", "cMo (2)", "PreRPM", "RPM"))
so_m$group <- "MoRPM (1) (NaCl) vs MoRPM (1) (CpG)"
```

```r
dea_m <- dea_seurat(so_m, ident="group",  only_pos=FALSE, logfc_threshold=0, min_pct=0, cnt_min=3, cell_min=3, conserved=FALSE, treatment=TRUE, grouping_var="treatment", test_use="wilcox", compute=TRUE)[[1]]
```

```r
options(repr.plot.width=10, repr.plot.height=7)

hm_treatment_m <- hm_treatment(so_m, dea_m, top=20, width=0.1, log2fc_thr=0.25, padj_thr=0.05, pct_thr=0.5, cell_type_fine=c("cMo (1)", "cMo (2)", "PreRPM", "RPM"), gene_filter=grep("Rps|Rpl", rownames(so_m), value=TRUE))
hm_treatment_m
```

```r
ggsave("result/figures/hm_treatment_m.pdf", hm_treatment_m, width=20, height=20, units="cm", dpi=1200)
```

```r
options(repr.plot.width=20, repr.plot.height=5)

dp_1 <- dp_dea_treatment(dea_m, so_m, log2fc_thr=0.25, padj_thr=0.05, type="receptor", group_by="cell_type_fine", group_by_order=c("cMo (1)", "cMo (2)", "PreRPM", "RPM")) + ggtitle("Receptor")
dp_2 <- dp_dea_treatment(dea_m, so_m, log2fc_thr=0.25, padj_thr=0.05, type="transcription_factor", group_by="cell_type_fine", group_by_order=c("cMo (1)", "cMo (2)", "PreRPM", "RPM")) + ggtitle("TF")
dp_3 <- dp_dea_treatment(dea_m, so_m, log2fc_thr=0.25, padj_thr=0.05, type="ligand", group_by="cell_type_fine", group_by_order=c("cMo (1)", "cMo (2)", "PreRPM", "RPM")) + ggtitle("Ligand")

dp_1 + dp_2 + dp_3 + plot_layout(ncol=3)
```

```r
options(repr.plot.width=10, repr.plot.height=5)

gsea_m_1 <- gsea_from_dea(dea_m, gene_sets=gene_set_hallmark)
gsea_plot_m_1 <- gsea_plot(gsea_m_1, pval_thr=0.1, pathway_suffix="HALLMARK", top=10)

gsea_m_2 <- gsea_m <- gsea_from_dea(dea_m, gene_sets=gene_set_go)
gsea_plot_m_2 <- gsea_plot(gsea_m_2, pval_thr=0.1, pathway_suffix="GOBP", top=10)

gsea_plot_m_1 + gsea_plot_m_2 
```

```r
options(repr.plot.width=7, repr.plot.height=2.5)

hm_treatment_m <- hm_treatment(so_m, dea_m[dea_m$gene %in% inf_genes, ], top=20, width=0.1, log2fc_thr=0.25, padj_thr=0.05, pct_thr=NULL, cell_type_fine=c("cMo (1)", "cMo (2)", "PreRPM", "RPM"))
hm_treatment_m
```

# DEA per cell type 

```r
so_m$group <- ifelse(so_m$cell_type_fine %in% c("RPM"), "RPM", "cMoPreRPM")
```

```r
dea_m <- dea_seurat(so_m, ident="group",  only_pos=FALSE, logfc_threshold=0, min_pct=0, cnt_min=3, cell_min=3, conserved=FALSE, treatment=TRUE, grouping_var="treatment", test_use="wilcox", compute=TRUE)
dea_m <- dea_m[c("cMoPreRPM", "RPM")]
```

```r
options(repr.plot.width=25, repr.plot.height=10)

hm_treatment_m <- lapply(names(dea_m), function(x) hm_treatment(subset(so_m, subset=group==x), dea_m[[x]], top=20, width=0.1, log2fc_thr=0.5, padj_thr=0.05, pct_thr=0.7, cell_type_fine=c("cMo (1)", "cMo (2)", "PreRPM", "RPM"), gene_filter=grep("Rps|Rpl", rownames(so_m), value=TRUE)))
hm_treatment_m <- hm_treatment_m[!unlist(lapply(hm_treatment_m, is.null))]
wrap_plots(hm_treatment_m, ncol=4)
```

```r
ggsave("result/figures/hm_treatment_cmoprerpm.pdf", hm_treatment_m[[1]], width=20, height=20, units="cm", dpi=1200)
```

```r
options(repr.plot.width=15, repr.plot.height=5)

gsea_m_1 <- lapply(dea_m, function(x) gsea_from_dea(x, gene_sets=gene_set_hallmark))
gsea_m_1 <- gsea_m_1[unlist(lapply(gsea_m_1, function(x) nrow(x)>0))]
gsea_plot_m_1 <- lapply(gsea_m_1, function(x) gsea_plot(x, pval_thr=1, pathway_suffix="HALLMARK", top=10) + theme_global_set())
                        
wrap_plots(gsea_plot_m_1, ncol=4)
```
