```r
gtf <- rtracklayer::import("/research/peer/fdeckert/reference/genome/GRCm38/gencode.vM25.annotation.gtf")
```

```r
genes <- c("Ly6a", "Adarb1")
```

```r
lapply(genes, function(i) {
    
    gn <- gtf[gtf$type=="gene" & gtf$gene_name==i]
    pr <- GenomicRanges::promoters(gn, upstream=100, downstream=500)
    rtracklayer::export(as.data.frame(gn)[, c("seqnames", "start", "end", "strand")], paste0("script/1_scRNAseq/8_tfbs/", tolower(i), ".bed"))

}
      )
```
